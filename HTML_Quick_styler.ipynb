```python
from IPython.display import HTML, display

# Simple, working Quick Styler for Colab
html_code = """
<!DOCTYPE html>
<html>
<head>
    <style>
        body {
            font-family: Arial, sans-serif;
            max-width: 1000px;
            margin: 0 auto;
            padding: 20px;
            background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
        }
        .container {
            background: white;
            border-radius: 20px;
            padding: 25px;
            box-shadow: 0 10px 40px rgba(0,0,0,0.2);
        }
        h1 {
            text-align: center;
            color: #333;
            margin-top: 0;
        }
        .subtitle {
            text-align: center;
            color: #666;
            margin-bottom: 30px;
        }
        .preview-area {
            background: #f5f5f5;
            border-radius: 15px;
            padding: 40px;
            margin-bottom: 30px;
            text-align: center;
            border: 2px dashed #ccc;
        }
        #preview-element {
            display: inline-block;
            padding: 20px 40px;
            background: #3498db;
            color: white;
            border-radius: 10px;
            font-size: 18px;
            transition: all 0.3s;
            box-shadow: 0 5px 15px rgba(0,0,0,0.2);
        }
        .controls-grid {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(250px, 1fr));
            gap: 20px;
            margin-bottom: 30px;
        }
        .control-card {
            background: #f8f9fa;
            border-radius: 12px;
            padding: 15px;
            border: 1px solid #e0e0e0;
        }
        .control-card h3 {
            margin-top: 0;
            color: #444;
            font-size: 16px;
            border-bottom: 2px solid #3498db;
            padding-bottom: 8px;
        }
        .slider-group {
            margin-bottom: 15px;
        }
        .slider-group label {
            display: block;
            margin-bottom: 5px;
            color: #555;
            font-size: 14px;
        }
        input[type=range] {
            width: 100%;
            margin: 5px 0;
        }
        .value-display {
            float: right;
            color: #3498db;
            font-weight: bold;
        }
        .color-group {
            display: flex;
            align-items: center;
            gap: 10px;
            margin-bottom: 12px;
        }
        .color-group span {
            min-width: 100px;
            color: #555;
        }
        .button-group {
            display: flex;
            gap: 10px;
            flex-wrap: wrap;
            margin-top: 10px;
        }
        .style-btn {
            flex: 1;
            padding: 8px;
            border: none;
            border-radius: 6px;
            background: #e0e0e0;
            color: #333;
            cursor: pointer;
            transition: all 0.2s;
            min-width: 70px;
        }
        .style-btn.active {
            background: #3498db;
            color: white;
        }
        .code-section {
            background: #2c3e50;
            border-radius: 12px;
            padding: 15px;
            margin-top: 20px;
        }
        .code-header {
            display: flex;
            justify-content: space-between;
            align-items: center;
            margin-bottom: 10px;
            color: white;
        }
        select {
            padding: 5px;
            border-radius: 5px;
            border: none;
        }
        pre {
            background: #34495e;
            color: #ecf0f1;
            padding: 15px;
            border-radius: 8px;
            overflow-x: auto;
            font-size: 13px;
            margin: 10px 0;
        }
        button {
            padding: 10px 20px;
            background: #3498db;
            color: white;
            border: none;
            border-radius: 8px;
            cursor: pointer;
            font-size: 14px;
            transition: all 0.2s;
        }
        button:hover {
            background: #2980b9;
        }
        .reset-btn {
            background: #e74c3c;
            margin-left: 10px;
        }
        .reset-btn:hover {
            background: #c0392b;
        }
        .footer {
            text-align: center;
            margin-top: 20px;
            color: #777;
            font-size: 13px;
        }
        .notification {
            position: fixed;
            top: 20px;
            right: 20px;
            background: #27ae60;
            color: white;
            padding: 10px 20px;
            border-radius: 8px;
            display: none;
        }
    </style>
</head>
<body>
    <div class="container">
        <h1>⚡ QUICK STYLER ADVANCED</h1>
        <div class="subtitle">Interactive CSS Styling Tool - Google Colab Edition</div>

        <!-- Preview Section -->
        <div class="preview-area">
            <div id="preview-element">
                ✨ QuickStyle Element
            </div>
        </div>

        <!-- Controls Section -->
        <div class="controls-grid">
            <!-- Spacing Card -->
            <div class="control-card">
                <h3>📏 Spacing</h3>
                <div class="slider-group">
                    <label>Padding <span id="padVal" class="value-display">20px</span></label>
                    <input type="range" id="padding" min="0" max="50" value="20">
                </div>
                <div class="slider-group">
                    <label>Border Radius <span id="radVal" class="value-display">10px</span></label>
                    <input type="range" id="radius" min="0" max="50" value="10">
                </div>
            </div>

            <!-- Typography Card -->
            <div class="control-card">
                <h3>🔤 Typography</h3>
                <div class="slider-group">
                    <label>Font Size <span id="fontVal" class="value-display">18px</span></label>
                    <input type="range" id="fontsize" min="12" max="36" value="18">
                </div>
                <div class="slider-group">
                    <label>Letter Spacing <span id="spaceVal" class="value-display">0px</span></label>
                    <input type="range" id="spacing" min="-2" max="8" value="0" step="0.5">
                </div>
            </div>

            <!-- Colors Card -->
            <div class="control-card">
                <h3>🎨 Colors</h3>
                <div class="color-group">
                    <span>Background</span>
                    <input type="color" id="bgColor" value="#3498db">
                </div>
                <div class="color-group">
                    <span>Text</span>
                    <input type="color" id="textColor" value="#ffffff">
                </div>
                <div class="color-group">
                    <span>Border</span>
                    <input type="color" id="borderColor" value="#2980b9">
                </div>
            </div>

            <!-- Border Style Card -->
            <div class="control-card">
                <h3>🔲 Border Style</h3>
                <div class="button-group">
                    <button class="style-btn active" id="styleSolid">Solid</button>
                    <button class="style-btn" id="styleDashed">Dashed</button>
                    <button class="style-btn" id="styleDotted">Dotted</button>
                    <button class="style-btn" id="styleNone">None</button>
                </div>
                <div style="margin-top: 15px;">
                    <button id="shadowBtn" class="style-btn">💡 Shadow: OFF</button>
                </div>
            </div>
        </div>

        <!-- Code Section -->
        <div class="code-section">
            <div class="code-header">
                <span>📋 Generated Code</span>
                <select id="codeType">
                    <option value="css">CSS</option>
                    <option value="html">HTML</option>
                    <option value="inline">Inline</option>
                </select>
            </div>
            <pre id="codeDisplay">/* Your CSS will appear here */</pre>
            <div style="text-align: right;">
                <button id="copyBtn">📋 Copy Code</button>
                <button id="resetBtn" class="reset-btn">↺ Reset All</button>
            </div>
        </div>

        <div class="footer">
            ⚡ Quick Styler v2.0 - Optimized for Google Colab
        </div>
    </div>

    <div class="notification" id="notification">✅ Code copied to clipboard!</div>

    <script>
        // Get elements
        const preview = document.getElementById('preview-element');
        const padding = document.getElementById('padding');
        const radius = document.getElementById('radius');
        const fontsize = document.getElementById('fontsize');
        const spacing = document.getElementById('spacing');
        const bgColor = document.getElementById('bgColor');
        const textColor = document.getElementById('textColor');
        const borderColor = document.getElementById('borderColor');

        // Value displays
        const padVal = document.getElementById('padVal');
        const radVal = document.getElementById('radVal');
        const fontVal = document.getElementById('fontVal');
        const spaceVal = document.getElementById('spaceVal');

        // Buttons
        const styleSolid = document.getElementById('styleSolid');
        const styleDashed = document.getElementById('styleDashed');
        const styleDotted = document.getElementById('styleDotted');
        const styleNone = document.getElementById('styleNone');
        const shadowBtn = document.getElementById('shadowBtn');
        const copyBtn = document.getElementById('copyBtn');
        const resetBtn = document.getElementById('resetBtn');
        const codeType = document.getElementById('codeType');
        const codeDisplay = document.getElementById('codeDisplay');
        const notification = document.getElementById('notification');

        // State
        let borderStyle = 'solid';
        let shadowEnabled = false;

        // Update function
        function updateAll() {
            // Update preview styles
            preview.style.padding = padding.value + 'px';
            preview.style.borderRadius = radius.value + 'px';
            preview.style.fontSize = fontsize.value + 'px';
            preview.style.letterSpacing = spacing.value + 'px';
            preview.style.backgroundColor = bgColor.value;
            preview.style.color = textColor.value;

            // Border
            if (borderStyle === 'none') {
                preview.style.border = 'none';
            } else {
                preview.style.border = '3px ' + borderStyle + ' ' + borderColor.value;
            }

            // Shadow
            if (shadowEnabled) {
                preview.style.boxShadow = '0 10px 25px rgba(0,0,0,0.3)';
                shadowBtn.textContent = '💡 Shadow: ON';
                shadowBtn.style.background = '#3498db';
                shadowBtn.style.color = 'white';
            } else {
                preview.style.boxShadow = 'none';
                shadowBtn.textContent = '💡 Shadow: OFF';
                shadowBtn.style.background = '#e0e0e0';
                shadowBtn.style.color = '#333';
            }

            // Update value displays
            padVal.textContent = padding.value + 'px';
            radVal.textContent = radius.value + 'px';
            fontVal.textContent = fontsize.value + 'px';
            spaceVal.textContent = spacing.value + 'px';

            // Update active border buttons
            [styleSolid, styleDashed, styleDotted, styleNone].forEach(btn => {
                btn.classList.remove('active');
            });

            if (borderStyle === 'solid') styleSolid.classList.add('active');
            if (borderStyle === 'dashed') styleDashed.classList.add('active');
            if (borderStyle === 'dotted') styleDotted.classList.add('active');
            if (borderStyle === 'none') styleNone.classList.add('active');

            // Update code display
            updateCode();
        }

        // Update code display
        function updateCode() {
            const type = codeType.value;

            if (type === 'css') {
                codeDisplay.textContent = '.quick-style-element {\n' +
                    '    padding: ' + padding.value + 'px;\n' +
                    '    border-radius: ' + radius.value + 'px;\n' +
                    '    font-size: ' + fontsize.value + 'px;\n' +
                    '    letter-spacing: ' + spacing.value + 'px;\n' +
                    '    background-color: ' + bgColor.value + ';\n' +
                    '    color: ' + textColor.value + ';\n' +
                    '    border: 3px ' + borderStyle + ' ' + borderColor.value + ';\n' +
                    '    box-shadow: ' + (shadowEnabled ? '0 10px 25px rgba(0,0,0,0.3)' : 'none') + ';\n' +
                    '}';
            } else if (type === 'html') {
                codeDisplay.textContent = '<div class="quick-style-element">\n' +
                    '    ✨ QuickStyle Element\n' +
                    '</div>';
            } else {
                codeDisplay.textContent = '<div style="padding: ' + padding.value + 'px; ' +
                    'border-radius: ' + radius.value + 'px; ' +
                    'font-size: ' + fontsize.value + 'px; ' +
                    'letter-spacing: ' + spacing.value + 'px; ' +
                    'background-color: ' + bgColor.value + '; ' +
                    'color: ' + textColor.value + '; ' +
                    'border: 3px ' + borderStyle + ' ' + borderColor.value + '; ' +
                    'box-shadow: ' + (shadowEnabled ? '0 10px 25px rgba(0,0,0,0.3)' : 'none') + ';">\n' +
                    '    ✨ QuickStyle Element\n' +
                    '</div>';
            }
        }

        // Event listeners
        padding.addEventListener('input', updateAll);
        radius.addEventListener('input', updateAll);
        fontsize.addEventListener('input', updateAll);
        spacing.addEventListener('input', updateAll);
        bgColor.addEventListener('input', updateAll);
        textColor.addEventListener('input', updateAll);
        borderColor.addEventListener('input', updateAll);

        // Border style buttons
        styleSolid.addEventListener('click', () => {
            borderStyle = 'solid';
            updateAll();
        });

        styleDashed.addEventListener('click', () => {
            borderStyle = 'dashed';
            updateAll();
        });

        styleDotted.addEventListener('click', () => {
            borderStyle = 'dotted';
            updateAll();
        });

        styleNone.addEventListener('click', () => {
            borderStyle = 'none';
            updateAll();
        });

        // Shadow toggle
        shadowBtn.addEventListener('click', () => {
            shadowEnabled = !shadowEnabled;
            updateAll();
        });

        // Code type change
        codeType.addEventListener('change', updateCode);

        // Copy button with fallback
        copyBtn.addEventListener('click', () => {
            const text = codeDisplay.textContent;

            // Try modern clipboard API first
            if (navigator.clipboard && navigator.clipboard.writeText) {
                navigator.clipboard.writeText(text).then(() => {
                    notification.style.display = 'block';
                    setTimeout(() => {
                        notification.style.display = 'none';
                    }, 2000);
                }).catch(() => {
                    // Fallback
                    fallbackCopy(text);
                });
            } else {
                // Fallback for older browsers
                fallbackCopy(text);
            }
        });

        // Fallback copy method
        function fallbackCopy(text) {
            const textarea = document.createElement('textarea');
            textarea.value = text;
            document.body.appendChild(textarea);
            textarea.select();
            try {
                document.execCommand('copy');
                notification.style.display = 'block';
                setTimeout(() => {
                    notification.style.display = 'none';
                }, 2000);
            } catch (err) {
                alert('Press Ctrl+C to copy');
            }
            document.body.removeChild(textarea);
        }

        // Reset button
        resetBtn.addEventListener('click', () => {
            padding.value = 20;
            radius.value = 10;
            fontsize.value = 18;
            spacing.value = 0;
            bgColor.value = '#3498db';
            textColor.value = '#ffffff';
            borderColor.value = '#2980b9';
            borderStyle = 'solid';
            shadowEnabled = false;
            updateAll();
        });

        // Initial update
        updateAll();
    </script>
</body>
</html>
"""

# Display in Colab
display(HTML(html_code))
```
